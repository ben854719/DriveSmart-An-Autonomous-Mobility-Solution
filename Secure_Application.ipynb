```python
import tenseal as ts
import numpy as np

# Normalize vectors for cosine similarity
def normalize(vec):
    norm = np.linalg.norm(vec)
    return [x / norm for x in vec]

# Inputs (replace with your actual embeddings)
input_embedding = [0.1, 0.3, 0.5, 0.7]
stored_embedding = [0.1, 0.3, 0.5, 0.7]
threshold = 0.9

# Normalize embeddings
input_embedding = normalize(input_embedding)
stored_embedding = normalize(stored_embedding)

# Create TenSEAL context
context = ts.context(
    ts.SCHEME_TYPE.CKKS,
    poly_modulus_degree=8192,
    coeff_mod_bit_sizes=[60, 40, 40, 60]
)
context.generate_galois_keys()
context.global_scale = 2 ** 40  # Recommended scale for CKKS

# Encrypt embeddings
enc_input = ts.ckks_vector(context, input_embedding)
enc_stored = ts.ckks_vector(context, stored_embedding)

# Compute encrypted dot product (cosine similarity)
enc_similarity = enc_input.dot(enc_stored)

# Decrypt result
similarity_score = enc_similarity.decrypt()[0]  # decrypt() returns a list

# Compare with threshold
is_match = similarity_score > threshold

# Output result
print(f"Similarity score: {similarity_score:.4f}")
print("Match:", is_match)
```
